## Cleaing Dataset Drop duplicated rows and columns


```python
# Drop duplicated rows
data.drop_duplicates(inplace=True)
# Reset index after dropping duplicates
data.reset_index(drop=True, inplace=True)
data.drop_duplicates(inplace=True)

data.shape
```

```python
data=data.drop(['Name','Club','Position','Begins','Expires'],axis=1)
data.head()
```

```python
# Columns in dataset table 
data.columns
```

```python
# strip all for leading/training white spaces 
data = data.apply(lambda x:x.str.strip() if x.dtype=='object' else x)
data.head(5)
```

## Converting Division into a category of top 5 or not 1 / 0

```python
# Converting Division into a category of top 5 or not 1 / 0
def is_top_5_league(league):

    top_5_league=['Premier League','Bundesliga','Serie A TIM','Ligue 1 Uber Eats','LALIGA EA SPORTS']

    if league in top_5_league:
        return 1
    else :
        return 0
df =pd.DataFrame()
df['IS_top_5_league']=data['Division'].apply(lambda x: is_top_5_league(x))
```

## Converting leagues' nations to rich nations with monetary power to pull players


```python
# Converting leagues' nations to rich nations with monetary power to pull players

def based_rich_nations(based):
  rich_nat = [
    "England",
    "China",
    "Germany",
    "Spain",
    "France",
    "Italy",
    "United States",
    "Qatar",
    "U.A.E.",
    "Saudi Arabia"]
  if based.split(' (')[0] in rich_nat:
    return 1
  else:
    return 0
      
df['Based_rich_nation']= data['Based'].apply(lambda x: based_rich_nations(x))
    
```

## Converting all nations into tiers of nations

```python
def is_top_ranked_nation(nation):
  tier1 = [ # top 10
    "ARG",  # Argentina
    "FRA",  # France
    "ENG",  # England
    "BEL",  # Belgium
    "BRA",  # Brazil
    "NED",  # Netherlands
    "POR",  # Portugal
    "ESP",  # Spain
    "ITA",  # Italy
    "CRO",  # Croatia
  ]

  tier2 = [ # 11-30
    "URU",  # Uruguay
    "MAR",  # Morocco
    "USA",  # USA
    "COL",  # Colombia
    "MEX",  # Mexico
    "GER",  # Germany 
    "SEN",  # Senegal
    "JPN",  # Japan
    "SUI",  # Switzerland
    "IRN",  # Iran
    "DEN",  # Denmark
    "KOR",  # Korea Republic
    "AUS",  # Australia
    "UKR",  # Ukraine
    "AUT",  # Austria
    "SWE",  # Sweden
    "HUN",  # Hungary
    "NGA",  # Nigeria
    "WAL",  # Wales
    "POL"]

  if nation in tier1:  # top 10
    return 2
  elif nation in tier2: # 11-30
    return 1
  else:
    return 0

df['Is_top_ranked_nation'] = data['Nat'].apply(lambda x: is_top_ranked_nation(x))
```

## Mapping EU National from YES/NO to 1/0

```python
# Mapping EU National from YES/NO to 1/0
df['EU_National'] = data['EU National'].map({'Yes': 1, 'No': 0})
```

##  Adding numerical data to our new df

```python
#@title Adding numerical data to our new df
df['Caps'] = data['Caps']
df['Apps'] = data['AT Apps'].replace('-', np.nan).fillna(0).astype(int)
df["Age"] = data["Age"]
df['Reputation'] = data['CR'].str.replace(',', '').astype(float)
```

## Checking to see if player's prev club was a top club which might indicate a higher wage

```python
## Checking to see if player's prev club was a top club which might indicate a higher wage
def is_top_prev_club(prev_club):
  # top 20 in past 10 yrs
  top_clubs = ['Real Madrid', 'Bayern München', 'Barcelona', 'Man City', 
               'Juventus', 'Chelsea', 'Atlético de Madrid', 'Liverpool', 
               'Paris Saint-Germain', 'Sevilla', 'Man Utd', 'Dortmund', 
               'Porto', 'Arsenal', 'Benfica', 'Ajax', 'Roma', 'Napoli', 'Tottenham', 'Shakhtar']
  if prev_club in top_clubs:
    return 1
  else:
    return 0

df['Is_top_prev_club'] = data['Last Club'].apply(lambda x: is_top_prev_club(x))
```

## Converting last transfer fee into numerical data

```python
# Converting last transfer fee into numerical data

def last_trans_fee_conversion(amt):
    # €203,478,000 p/a  203478000
    if amt == '-' or '€0':
        return 0
    elif amt[-1] == 'M' or 'K':
        return float(amt[1:-1])

df['Last_Transfer_Fee'] = data['Last Trans. Fee'].apply(lambda x: last_trans_fee_conversion(x))
```

## Converting Salary into numerical data


```python
## Converting Salary into numerical data
# €48,467,000 p/a
def salary_conversion(amt):
  return float(amt[1:].split()[0].replace(',', ''))

df['Salary'] = data['Salary'].apply(lambda x: salary_conversion(x))
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
columns_to_scale = ["Caps", "Apps", "Age", "Reputation"]
df[columns_to_scale] = sc.fit_transform(df[columns_to_scale])
df.head()
```
